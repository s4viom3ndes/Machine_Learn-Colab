# Abrindo o Arquivo .CSV
## Este conjunto de dados tem como objetivo classificar a probabilidade de existência de células cancerígenas com base em informações relacionadas às características de tumores, como área superficial, rugosidade, volume, entre outros.
## Utilizamos algoritmos de classificação para identificar quais tumores têm potencial carcinogênico, com base nos dados fornecidos no arquivo.

```python
import pandas as pd
import numpy


dataset = pd.read_csv("/content/drive/MyDrive/Cancer_Data.csv", sep = ",", encoding= "iso-8859-1")
dataset
```

```python
dataset.shape
```

# Excluindo a coluna: "ID"

```python
dataset_01 = dataset

dataset_01 = dataset_01.drop(columns=['id'])
dataset_01
```

# Relação entre trumores benignos e malignos

```python
import seaborn as sns
dataset_01.value_counts("diagnosis")
```

```python
sns.countplot(x = "diagnosis",  data = dataset_01)
```

# Transformando os valores da coluna "diagnosis" para valores numericos: 0 e 1

```python
dataset_02 = dataset_01

dataset_02['diagnosis'] = dataset_02['diagnosis'].replace('M', 1, regex=True) # Com câncer
dataset_02['diagnosis'] = dataset_02['diagnosis'].replace('B', 0, regex=True) # Sem câncer

dataset_02
```

# Usando KNN

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix


x = dataset_02.iloc[:, 1:-1].values # matriz que contem os dados a serem cruzados
y = dataset_02.iloc[:, 0].values # valor alvo (B ou M)

```

```python
from sklearn.neighbors import KNeighborsClassifier

knn = KNeighborsClassifier(n_neighbors = 5, metric="minkowski", p = 2)

modelo = knn.fit(x,y)
```

```python
modelo
```

```python
y_prev = modelo.predict(x)
print(y_prev)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(8, 6))
plt.scatter(x[:, 0],x[:, 1], c=y_prev, cmap='viridis')
plt.xlabel('diagnosis')
plt.ylabel('radius_mean')
plt.title('Previsões do Modelo KNN')
plt.colorbar(label='Classes Previstas')
plt.show()
```

```python
acerto_mizeravi = accuracy_score(y, y_prev)
precision = precision_score(y, y_prev, average = 'weighted')
recall = recall_score(y, y_prev, average='weighted')
f1 = f1_score(y, y_prev, average='weighted')
print(f'acerto, mizeravi ? : {acerto_mizeravi}, precisao: {precision}, Recall {recall}, F1: {f1}')

matriz = confusion_matrix(y, y_prev)
matriz
```

# Agora utilizando um arvore de decisão, iremos prever os resultados e calcular uma nova acurácia

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import export_graphviz
import graphviz
from sklearn.tree import  plot_tree
```

```python
x_treinamento, x_teste, y_treinamento, y_teste = train_test_split(x, y, test_size=0.3, random_state=1)

model = RandomForestClassifier(random_state=1, n_estimators=100, max_depth=50, max_leaf_nodes=10)
model.fit(x_treinamento, y_treinamento)

```

```python
previsoes = model.predict(x_teste)
indice = 0
arvore = model.estimators_[indice]
plt.figure(figsize=(20,10))
colunas = dataset_02.columns[:-1].tolist()
plot_tree(arvore, filled = True, feature_names = colunas , rounded=True )
plt.show
```

```python
acerto_mizeravi = accuracy_score(y_teste, previsoes)
precision = precision_score(y_teste, previsoes, average = 'weighted')
recall = recall_score(y_teste, previsoes, average='weighted')
f1 = f1_score(y_teste, previsoes, average='weighted')
print(f'acerto, mizeravi ? : {acerto_mizeravi}, precisao: {precision}, Recall {recall}, F1: {f1}')
```
