```python
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.preprocessing import LabelEncoder
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, classification_report
from sklearn.tree import export_graphviz
import graphviz

import matplotlib.pyplot as plt
from sklearn.tree import  plot_tree
```

```python
dataframe = pd.read_csv("/content/titanic.csv")
dataframe
```

```python
x = dataframe.iloc[:, [1, 3, 4, 5]].values
y = dataframe.iloc[:, 0].values



x_transformed = x.copy()



label_encoder = LabelEncoder()
for i in range(x.shape[1]):
    if x[:, i].dtype == "object":
        x_transformed[:, i] = label_encoder.fit_transform(x[:, i])


x_treinamento, x_teste, y_treinamento, y_teste = train_test_split(x_transformed, y, test_size=0.3, random_state=1)

modelo = RandomForestClassifier(random_state=1, n_estimators=100, max_depth=50, max_leaf_nodes=10)
modelo.fit(x_treinamento, y_treinamento)



```

```python
previsoes = modelo.predict(x_teste)
```

```python
indice = 0
arvore = modelo.estimators_[indice]
plt.figure(figsize=(20,10))
plot_tree(arvore, filled = True, feature_names = dataframe.columns[:-1].tolist(), rounded=True )
plt.show
```

```python
acerto_mizeravi = accuracy_score(y_teste, previsoes)
precision = precision_score(y_teste, previsoes, average = 'weighted')
recall = recall_score(y_teste, previsoes, average='weighted')
f1 = f1_score(y_teste, previsoes, average='weighted')
print(f'acerto, mizeravi ? : {acerto_mizeravi}, precisao: {precision}, Recall {recall}, F1: {f1}')
```
